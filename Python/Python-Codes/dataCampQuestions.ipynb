```python
# Se te asigna un DataFrame con las columnas y como cadenas.
# data datetime 
# crea una columna como tipo de dato fecha y hora combinando las dos columnas. datetime
# DataFrame: un dataframe es una estructura de datos bidimensional, similar a una tabla, que se utiliza en Python para almacenar y manipular datos de manera eficiente. Es proporcionado por la biblioteca pandas y permite realizar operaciones como filtrado, agrupamiento y análisis de datos de manera sencilla.
"""  
Fecha       Hora
2023-07-01  12:30 PM
2023-07-02  03:45 PM
2023-07-03  09:15 AM
"""

import pandas as pd

# Creacion del DataFrame

data = pd.DataFrame({
    'Fecha': ['2023-07-01', '2023-07-02', '2023-07-03'],
    'Hora': ['12:30 PM', '03:45 PM', '09:15 AM']
})

data['datetime'] = pd.to_datetime(data['Fecha'] + ' ' + data['Hora'])
print(data['datetime'])
# pd.to_datetime() es una funcion de pandas que dectecta automaticamente el formato de fecha y hora y convierte las cadenas en objetos datetime de pandas.
# data --> nombre del DataFrame 
# ['datetime'] --> nombre de la columna entre corchetes (como un diccionario)
```

```python
# El conjunto de datos dado tiene precios para diferentes modelos de telefono. df
# Busca todos los modelos que contienen la cadena "Pro" en el usando regex. Model

"""  
Marca    Modelo          Precio
Apple    iPhone 13 Pro   999
Samsung  Galaxy S21 Pro  799
Google   Pixel 6         599
OnePlus  9 Pro           729

"""

import re # importar la libreria de expresiones regulares --> herramienta que permite trabajar con patrones de texto de manera poderosa y flexible.
import pandas as pd

data = pd.DataFrame({
    'Marca': ['Apple', 'Samsung', 'Google', 'OnePlus'],
    'Modelo': ['iPhone 13 Pro', 'Galaxy S21 Pro', 'Pixel 6', '9 Pro'],
    'Precio': [999, 799, 599, 729]
})

pattern = r'Pro'  # cadena a buscar en el modelo usando regex
filtered_models = data[data['Modelo'].str.contains(pattern, regex=True)]
print(filtered_models)

"""  
Aclaraciones:

r --> indica que es una cadena raw (cruda), lo que significa que los caracteres de escape se interpretan literalmente.

"Pro" --> cadena que se busca en los modelos de telefono.


.str.contains() --> metodo de pandas que verifica si una cadena contiene un patron especifico (en este caso, "Pro").

refex=True --> indica que se esta utilizando una expresion regular para la busqueda.

"""
```

```python
# Del conjunto de datos dado, filtra el conjunto de datos para mostrar solo filas donde empieza con la letra "T". df restaurant 

"""

Restaurante             Gastronomia
The Gourmet Kitchen     Italian
Hora del sushin         Japanese
Taco Fiesta             Mexican
restaurante de curry     Indian

"""

import pandas as pd

data = pd.DataFrame({
    'Restaurante': ['The Gourmet Kitchen', 'Hora del sushin', 'Taco Fiesta', 'restaurante de curry'],
    'Gastronomia': ['Italian', 'Japanese', 'Mexican', 'Indian']
})

filtered_restaurants = data.loc[data['Restaurante'].str.startswith('T')]
print(filtered_restaurants)

# data.loc[] --> metodo de pandas que permite filtrar filas y columnas basadas en etiquetas o condiciones especificas.
# .str.startswith('T') --> metodo de pandas que verifica si una cadena comienza con un patron especifico (en este caso, la letra "T").
```

```python
# Dado un dataframe pandas con columna, convierte el a formato Mes/Dia/Año. Date

"""  
Fecha       Region
2023-07-01  Un
2023-08-15  B
2023-09-30  C

"""

import pandas as pd 

data = pd.DataFrame({
    'Fecha': ['2023-07-01', '2023-08-15', '2023-09-30'],
    'Region': ['A', 'B', 'C']
})

data['Fecha'] = pd.to_datetime(data['Fecha']).dt.strftime('%m/%d/%Y')
print(data)
```

```python
# dada una cadena llamada html que contiene una tabla con columnas y lee el script html y carga la tabla como 
# un dataframe pandas


import pandas as pd

html_script = """
<table>
    <tr>
        <th>Nombre</th>
        <th>Edad</th>
        <th>Ciudad</th>
    </tr>
    <tr>
        <td>Alice</td>
        <td>30</td>
        <td>New York</td>
    </tr>
    <tr>
        <td>Bob</td>
        <td>25</td>
        <td>Los Angeles</td>
    </tr>
    <tr>
        <td>Charlie</td>
        <td>35</td>
        <td>Chicago</td>
    </tr>
</table>
"""

# leer con html con conversion de tipos

html_parsed = pd.read_html(html_script, converters={
    'Edad': int,
    'Nombre': str,
    'Ciudad': str,
    })

# obtener el dataframe (read_html devuelve una lista de dataframes)
df = html_parsed[0]

print(df)
```

```python
# Cuenta el numero de filas en la tabla de peliculas de la base de datos de prueba


import sqlalchemy as bd

engine = bd.create_engine("sqlite:///test_db.sqlite")
connection = engine.connect()

query = connection.execute("SELECT COUNT(*) FROM movies") # consulta SQL para contar el numero de filas en la tabla movies
result = query.fetchone() # obtener el resultado de la consulta
count = result.fetchone()[0] # extraer el numero de filas del resultado
print("Número de filas en la tabla de películas: ", count)

conn.close()

# Aclaraciones:
# engine --> objeto que representa la conexion a la base de datos
# connection --> objeto que representa la conexion activa a la base de datos
# query --> objeto que representa la consulta SQL ejecutada
# result --> objeto que contiene el resultado de la consulta
# count --> variable que almacena el numero de filas en la tabla de peliculas

# EJEMPLO MAS COMPLETO OCUPADO UNA BASE DE DATOS REAL EN ESTA MISMA CARPETA :)
```
